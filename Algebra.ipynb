```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([[2, -1], [-1, 2]])
b = np.array([0, 3])
```

```python
#Solving using Numpy Solver
x = np.linalg.solve(a, b)
```

```python
#Solving using Numpy matrix inverse
ainv = np.linalg.inv(a)
x2 = np.linalg.inv(a) @ b
```

```python
print("X solved using Numpy Solver:         {}\nX solved using Numpy matrix inverse: {}".format(x,x2))
```

```python
va1 = a[:,0]    # First column vector of a
va2 = a[:,1]    # Second column vector of a
```

```python
x_scaled_1 = (va1 * x[0])     # First column vector of a scaled by X[0]
x_scaled_2 = (va2 * x[1])     # Second column vector of a scaled by X[1]
```

```python
# Plotting Column vectors
plt.plot([0, va1[0]], [0,va1[1]], label='First column Vector A')
plt.plot([0, va2[0]], [0,va2[1]], label='Second column Vector A')

# Plotting scaled column vectors
plt.plot([0, x_scaled_1[0]], [0,x_scaled_1[1]], label='VA1 Scaled by X[0]')
plt.plot([x_scaled_1[0], x_scaled_1[0]+x_scaled_2[0]], [x_scaled_1[1],x_scaled_1[1]+x_scaled_2[1]], label='VA2 Scaled by X[1]')

# Recalculating B and plotting it
b_recalculated = x_scaled_1 + x_scaled_2
plt.plot([0, b_recalculated[0]], [0, b_recalculated[1]], label='Linear Combination')

plt.title('Linear Combination Process')
plt.legend()
plt.grid()
plt.show()
```
