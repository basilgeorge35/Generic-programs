```python
import numpy as np
import matplotlib.pyplot as plt

#Function to emulate Binomial trials
def binomial(n, p, trials):
    successes = np.zeros(trials, dtype=int)     #Array to store successful result of each trial
    for i in range(trials):
        draws = np.random.choice(['red', 'black'], size=n, p=[1-p, p])  #Randomly picking n=10 balls with p=0.3 (30% probability) for Red and p= 0.7 (70%) for Black
        successes[i] = np.sum(draws == 'black')                         #Recording number of Black balls in each trial
    return successes
```

```python
n = 10              #Number of picks in each trial
p = 0.7             #Probability of picking Black ball
trials = 10000      #Number of trials

successes = binomial(n, p, trials)

plt.hist(successes, color='b', alpha = 0.6)
plt.xlabel('Successes')
plt.ylabel('Probability')
plt.show()
```
